```python
folder = "../data/index"
repo = "pandas"
file = "commitsIndex.ga"
path = "{}/{}/{}".format(folder, repo, file)
```

```python
dataset = []
with open(path, "r") as inp:
    for line in inp.readlines():
        files = line.strip().split()[1:]
        dataset.append(files)
dataset = dataset[:500000]
dataset
```

```python
patterns = pyfpgrowth.find_frequent_patterns(dataset, len(dataset)*0.001)
```

```python
patterns
```

```python
rules = pyfpgrowth.generate_association_rules(patterns, 0.5)
rules
```

```python
sorted_rules = sorted(rules.items(), key=lambda value: -value[1][1])
```

```python
sorted_rules
```

```python
log = open("{}.log".format(repo), "w")
st_to_print = ""
for rule in sorted_rules:
    from_files = ' '.join(rule[0])
    to_files = ' '.join(rule[1][0])
    score = rule[1][1]
    st_to_print += "{}|{}|{}\n".format(from_files, to_files, score)
print(st_to_print, file=log)
log.close()
st_to_print
```
